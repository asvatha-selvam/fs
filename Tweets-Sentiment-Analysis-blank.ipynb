```python
import re
import tweepy
from tweepy import OAuthHandler
from textblob import TextBlob

class TwitterClient(object):
    '''
    Generic Twitter Class for sentiment analysis.
    '''
    def __init__(self):
        '''
        Class constructor or initialization method.
        '''
        # keys and tokens from the Twitter Dev Console
        consumer_key = 'ts8RhO1B1qIzdSLaDsaise89a'
        consumer_secret = 'A4ZawHMrKiAaoxfQhmjrW5F2gWwgjh9jhCuvczy8cymLaTJfES'
        access_token = '929606650415136768-Ok0ca8lkQk7oD8xNUqKWKg9VyoEXJRV'
        access_token_secret = '5mJm07yiPpvFB7M4Z9l4yaip4k6gvSpL2mEMMs6sEXjqX'

        # attempt authentication
        try:
            # create OAuthHandler object
            self.auth = OAuthHandler(consumer_key, consumer_secret)
            # set access token and secret
            self.auth.set_access_token(access_token, access_token_secret)
            # create tweepy API object to fetch tweets
            self.api = tweepy.API(self.auth)
        except:
            print("Error: Authentication Failed")

    def clean_tweet(self, tweet):
        '''
        Utility function to clean tweet text by removing links, special characters
        using simple regex statements.
        '''
        return ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", tweet).split())

    def get_tweet_sentiment(self, tweet):
        '''
        Utility function to classify sentiment of passed tweet
        using textblob's sentiment method
        '''
        # create TextBlob object of passed tweet text
        analysis = TextBlob(self.clean_tweet(tweet))
        # set sentiment
        if analysis.sentiment.polarity > 0:
            return 'positive'
        elif analysis.sentiment.polarity == 0:
            return 'neutral'
        else:
            return 'negative'

    def get_tweets(self, query, count=10):
        '''
        Main function to fetch tweets and parse them.
        '''
        # empty list to store parsed tweets
        tweets = []

        try:
            # call twitter api to fetch tweets
            fetched_tweets = self.api.search_tweets(q=query, count=count)

            # parsing tweets one by one
            for tweet in fetched_tweets:
                # empty dictionary to store required params of a tweet
                parsed_tweet = {}

                # saving text of tweet
                parsed_tweet['text'] = tweet.text
                # saving sentiment of tweet
                parsed_tweet['sentiment'] = self.get_tweet_sentiment(tweet.text)

                # appending parsed tweet to tweets list
                if tweet.retweet_count > 0:
                    # if tweet has retweets, ensure that it is appended only once
                    if parsed_tweet not in tweets:
                        tweets.append(parsed_tweet)
                else:
                    tweets.append(parsed_tweet)

            # return parsed tweets
            return tweets

        except tweepy.TweepError as e:
            # print error (if any)
            print("Error : " + str(e))

def main():
    # creating object of TwitterClient Class
    api = TwitterClient()
    # calling function to get tweets
    tweets = api.get_tweets(query='Donald Trump', count=200)

    # picking positive tweets from tweets
    ptweets = [tweet for tweet in tweets if tweet['sentiment'] == 'positive']
    # percentage of positive tweets
    print("Positive tweets percentage: {} %".format(100 * len(ptweets) / len(tweets)))
    # picking negative tweets from tweets
    ntweets = [tweet for tweet in tweets if tweet['sentiment'] == 'negative']
    # percentage of negative tweets
    print("Negative tweets percentage: {} %".format(100 * len(ntweets) / len(tweets)))
    # percentage of neutral tweets
    print("Neutral tweets percentage: {} %".format(100 * (len(tweets) - (len(ntweets) + len(ptweets))) / len(tweets)))

    # printing first 10 positive tweets
    print("\n\nPositive tweets:")
    for tweet in ptweets[:10]:
        print(tweet['text'])

    # printing first 10 negative tweets
    print("\n\nNegative tweets:")
    for tweet in ntweets[:10]:
        print(tweet['text'])

if __name__ == "__main__":
    # calling main function
    main()
```

```python
import instaloader
import numpy as np
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
import tensorflow as tf

# Initialize Instaloader
L = instaloader.Instaloader()
# Log in to Instagram (replace 'your_username' and 'your_password' with your actual credentials)
L.login('asvatha_selvam', 'password')

# Define a function to fetch post captions from a user's profile
def fetch_post_captions_and_comments(username):
    captions_and_comments = []
    profile = instaloader.Profile.from_username(L.context, username)
    for post in profile.get_posts():
        post_data = {}
        if post.caption:
            post_data['caption'] = post.caption
        comments = []
        for comment in post.get_comments():
            comments.append(comment.text)
        post_data['comments'] = comments
        captions_and_comments.append(post_data)
    return captions_and_comments

# Fetch post captions and comments from a user's profile
username = '_kvn_ks_'
posts_data = fetch_post_captions_and_comments(username)

# Tokenize the captions and comments
tokenizer = Tokenizer()
all_text = []
for post_data in posts_data:
    if 'caption' in post_data:
        all_text.append(post_data['caption'])
    all_text.extend(post_data['comments'])

tokenizer.fit_on_texts(all_text)
sequences = tokenizer.texts_to_sequences(all_text)

# Pad sequences to ensure uniform length
max_len = 100  # Desired sequence length
padded_sequences = pad_sequences(sequences, maxlen=max_len, padding='post')

# Load the trained LSTM model
model = tf.keras.models.load_model('lstm_model.h5')

# Predict sentiment using the model
sentiment_probabilities = model.predict(padded_sequences)

# Convert probabilities to predicted classes
sentiment_predictions = np.argmax(sentiment_probabilities, axis=1)

# Map predictions to sentiment labels
sentiment_labels = {0: 'Positive', 1: 'Negative', 2: 'Neutral'}
predicted_sentiments = [sentiment_labels[pred] for pred in sentiment_predictions]

# Print predicted sentiments for each post's caption and comments
for i, post_data in enumerate(posts_data, start=1):
    print(f"Post {i}:")
    if 'caption' in post_data:
        print(f"Caption: {post_data['caption']}\nPredicted Sentiment (Caption): {predicted_sentiments[i-1]}")
    if 'comments' in post_data:
        for j, comment in enumerate(post_data['comments'], start=1):
            print(f"Comment {j}: {comment}\nPredicted Sentiment (Comment): {predicted_sentiments[len(posts_data) + j - 1]}")
    print()
```
